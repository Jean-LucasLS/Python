```python
import pandas as pd

PATH = 'https://raw.githubusercontent.com/adamerose/datasets/master/titanic.csv'
df   = pd.read_csv(PATH)
```

```python
df.tail()
```

```python
df.fillna({'age': df['age'].mean(), 'deck': 'Unkown'}).tail()
```

```python
(
  df
  .assign(
    age_bfill       = df['age'].bfill(),
    age_ffill       = df['age'].ffill(),
    age_interpolate = df['age'].interpolate()
  )
  .filter(like='age')
  .tail()
  .style.highlight_null('purple')
)
```

```python
df.groupby('sex')['age'].apply(lambda x: x.fillna(x.mean())) # Os valores faltantes das idades serão preenchidas com as médias de cada sexo
```
