```python
int_idx = df['Total Charges'].map(lambda x: isinstance(x, int))
str_idx = df['Total Charges'].map(lambda x: isinstance(x, str))

df.loc[int_idx, 'Total Charges'] = df.loc[int_idx, 'Total Charges'].astype(float)
df.loc[str_idx, 'Total Charges'] = 0
```

```python
df['Total Charges'] = df['Total Charges'].astype(float)
```

```python
df[df['Total Charges']==0]
```

```python
df = df[df['Total Charges']!=0]
```

```python
num_cols = df.select_dtypes('number').columns

df[num_cols].apply(scale).plot.box(figsize=(18,7)); 
```

```python
df_features = df.drop(columns='Churn Value')
df_target   = df['Churn Value']
```

```python
num_cols   = df_features.select_dtypes('number').columns
categ_cols = df_features.select_dtypes(exclude='number').columns
```

```python
preprocessing = ColumnTransformer([
  ('numerical_preprocessing', StandardScaler(), num_cols),
  ('categorical_preprocessing', OneHotEncoder(), categ_cols),
])

preprocessing
```

```python
preprocessing.fit(df_features);
```

```python
pd.DataFrame(
  data=preprocessing.transform(df_features),
  columns=preprocessing.get_feature_names_out()
)
```

```python
df_mi = df.copy()

encoder = OrdinalEncoder()
df_mi[categ_cols] = encoder.fit_transform(df_mi[categ_cols]).astype(int)
df_mi.head()
```

```python
from sklearn.feature_selection import mutual_info_classif

df_features = df_mi.drop(columns='Churn Value')
target      = df_mi['Churn Value']
disc_cols   = df_features.dtypes==int

mi_scores   = mutual_info_classif(X=df_features, y=target, discrete_features=disc_cols)
mi_scores
```

```python
pd.Series(mi_scores, index=df_features.columns).sort_values().plot.barh(figsize=(16,8))
plt.xlabel('MI Scores')
plt.show()
```
