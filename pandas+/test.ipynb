```python
import matplotlib.pyplot as plt

categorias = ['Categoria 1', 'Categoria 2', 'Categoria 3', 'Categoria 4']
valores = [10, 20, 15, 25]

fig, ax = plt.subplots()
ax.bar(categorias, valores)

ax.set_title('Gr√°fico de Barras')
ax.set_xlabel('Categorias')
ax.set_ylabel('Valores')

plt.show()
```
