```python
from sklearn import datasets

df,_ = datasets.load_wine(return_X_y=True, as_frame=True)
print(df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

df.plot.box()
plt.xticks(rotation=60, ha='right');
```

```python
from sklearn.preprocessing import scale

df.apply(scale).plot.box() # A função apply(scale) facilita a visualização de outliers no gráfico de box(), pois aproxima a escala das colunas observadas
plt.xticks(rotation=60, ha='right');
```
