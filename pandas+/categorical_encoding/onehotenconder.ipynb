```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('tips')
df.head()
```

```python
pd.get_dummies(df) # Tratativa para variáveis categóricas (no pandas)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='if_binary')
encoder

OneHotEncoder(drop='if_binary')

categorical_columns = df.select_dtypes(exclude='number').columns

encoder.fit(df[categorical_columns])

encoder.categories_ # Mostra a ordem de categorias por coluna
encoder.drop_idx_
encoder.feature_names_in_ # Categorias entradas no modelo
encoder.get_feature_names_out() # Como ficará depois de aplicado o modelo

encoder.transform(df[categorical_columns]).toarray()
```

```python
pd.DataFrame(
  data = encoder.transform(df[categorical_columns]).toarray(),
  columns = encoder.get_feature_names_out()
)
```
