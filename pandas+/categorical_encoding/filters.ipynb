```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('tips')
df.head()
```

```python
df[df.tip > 4].head()
```

```python
df.query('tip > 4').head() # Utilização da query
```

```python
minimum = 3
df.query('tip > @minimum') # Utilização de variáveis dentro da query '@variable'
```

```python
maximum_bill = 20
minimum_tip  = 3
df.query('tip > @minimum_tip & total_bill < @maximum_bill').head() # Múltiplos filtros dentro de uma query
```
