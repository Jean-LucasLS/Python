```python
# Identificar quantos valores duplicados há em cada linha duplicada
(
  df
  .groupby(df.columns.tolist(), dropna=False) # Não irá ignorar os valores faltantes
  .size()
  .to_frame('n_duplicates')
  .query('n_duplicates>1') # Buscar valores que aparecem mais de uma vez
  .sort_values('n_duplicates', ascending=False)
  .head(8)
)
```
