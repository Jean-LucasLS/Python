```python
# A função 'concat' do pandas funciona de maneira similar ao 'union' da linguagem SQL

import pandas as pd

df1 = pd.DataFrame(
  {
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
  }, index=range(4)
)

df2 = pd.DataFrame(
  {
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'D': ['D4', 'D5', 'D6', 'D7'],
    'F': ['F4', 'F5', 'F6', 'F7']
  }, index=(4, 5, 6, 7)
)

df3 = pd.DataFrame(
  {
    'A': ['A8', 'A9', 'A10', 'A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'D': ['D8', 'D9', 'D10', 'D11']
  }, index=range(8, 12, 1)
)

df4 = pd.DataFrame(
  {
    'B': ['B2', 'B3', 'B6', 'B7'],
    'D': ['D2', 'D3', 'D6', 'D7'],
    'F': ['F2', 'F3', 'F6', 'F7']
  }, index=(2, 3, 6, 7)
)
```

```python
from IPython.display import display_html

def display_sbs(dataframes:list, titles:list):
  html_str=''
  for df, title in zip(dataframes, titles):
    html_str+=f'<span>{title}:</span>'
    html_str+=df.to_html().replace('table', 'table style="display:inline"')
    html_str+='&nbsp'*10
  display_html(html_str,raw=True)
```

```python
display_sbs([df1, df4, pd.concat([df1, df4], axis=1)], ['df1', 'df4', 'concat'])
```

```python
df_A = df1.iloc[:2] # Pega as duas primeiras linhas
df_B = df1.iloc[2:] # Pega as duas últimas linhas

display_sbs([df_A, df_B], ['df_A', 'df_B'])
```

```python
pd.concat([df_A, df_B])
```

```python
pd.concat([df_A, df_B], axis=1)
```

```python
pd.concat([df_A, df_B.reset_index(drop=True)], axis=1,) # Reset dos índices do DataFrame 'df_B'
```
