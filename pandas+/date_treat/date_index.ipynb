```python
import pandas as pd

PATH       = 'https://query.data.world/s/2gfb7bmzhna6kcbpc7admwa6cexprz'
df         = pd.read_csv(PATH, usecols=['Date', 'Fatalities']).dropna()
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y') # O argumento passado em 'format=' garante que a data será lida e convertida de forma correta

print(df.shape)
df.tail()
```

```python
df_index = df.set_index('Date').sort_index() # Adiciona a data como 'index' e as ordena
df_index.loc['2000-05-10':'2003-08'] # Filtro de um intervalo de data
```

```python
df_index.resample('M').sum() # Soma de 'fatalities' agrupadas por mês ('M')
```
