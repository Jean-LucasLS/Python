```python
import matplotlib.pyplot as plt

# Dados de exemplo
dados1 = [1, 2, 3, 4, 5]
dados2 = [2, 4, 6, 8, 10]

# Criação do gráfico box com duas caixas lado a lado
plt.boxplot([dados1, dados2], labels=['Grupo 1', 'Grupo 2'], patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            medianprops=dict(color='red'))

# Adiciona um título ao gráfico
plt.title('Exemplo de Gráfico Box')

# Adiciona rótulos aos eixos
plt.xlabel('Grupos')
plt.ylabel('Valores')

# Exibição do gráfico
plt.show()
```
