```python
import pandas as pd

PATH = './USA_Housing.csv'
df   = pd.read_csv(PATH)

df.info()
```

```python
import seaborn as sns
import numpy as np
import chart_studio.plotly as py

import plotly.graph_objects as go
import plotly.express as px

# A plotagem de gráficos box, ou boxplot, é uma forma eficaz de visualizar a distribuição de um conjunto de dados e identificar valores atípicos.
fig = px.box(df, x='Avg. Area Income')
fig.update_traces(line_color='blue')
```

```python
df = df.rename(columns={'Avg. Area Income': 'Avg_Area_Income', 'Avg. Area House Age': 'Avg_Area_House_Age',
                        'Avg. Area Number of Rooms': 'Avg_Area_Number_of_Rooms', 'Avg. Area Number of Bedrooms':
                        'Avg_Area_Number_of_Bedrooms', 'Area Population': 'Area_Population'})
df = df.drop(columns=['Address'])

df.head()
```

```python
'''A função 'pairplot' tem como objetivo traçar um conjunto de gráficos de dispersão entre pares de variáveis em um DataFrame. Essa função cria uma matriz de gráficos que
mostra as relações bivariadas entre diferentes variáveis, permitindo uma visualização rápida e conveniente da distribuição dos dados e das relações entre as variáveis.'''

# sns.pairplot(df)
sns.pairplot(df, x_vars=['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number_of_Rooms', 'Avg_Area_Number_of_Bedrooms', 'Area_Population'], y_vars='Price')
```

```python
'''A função corr() é uma função estatística utilizada para calcular a correlação entre duas variáveis. A correlação
mede a relação entre as variáveis e indica a força e a direção dessa relação. Ela é amplamente utilizada em análises
estatísticas para determinar o grau de associação entre as variáveis e entender se existe uma relação linear entre elas.

• A função corr() retorna um valor que varia de -1 a 1, em que:
  → Um valor de 1 indica uma correlação positiva perfeita, ou seja, as variáveis estão perfeitamente correlacionadas de forma positiva.
  → Um valor de -1 indica uma correlação negativa perfeita, ou seja, as variáveis estão perfeitamente correlacionadas de forma negativa.
  → Um valor próximo a 0 indica uma correlação fraca ou inexistente entre as variáveis.'''
  
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True)
```
