```python
import pandas as pd

PATH = './USA_Housing.csv'
df   = pd.read_csv(PATH)

df.info()
```

```python
import seaborn as sns
import numpy as np
import chart_studio.plotly as py

import plotly.graph_objects as go
import plotly.express as px

# A plotagem de grficos box, ou boxplot,  uma forma eficaz de visualizar a distribuio de um conjunto de dados e identificar valores atpicos.
fig = px.box(df, x='Avg. Area Income')
fig.update_traces(line_color='blue')
```

```python
df = df.rename(columns={'Avg. Area Income': 'Avg_Area_Income', 'Avg. Area House Age': 'Avg_Area_House_Age',
                        'Avg. Area Number of Rooms': 'Avg_Area_Number_of_Rooms', 'Avg. Area Number of Bedrooms':
                        'Avg_Area_Number_of_Bedrooms', 'Area Population': 'Area_Population'})
df = df.drop(columns=['Address'])

df.head()
```

```python
'''A funo 'pairplot' tem como objetivo traar um conjunto de grficos de disperso entre pares de variveis em um DataFrame. Essa funo cria uma matriz de grficos que
mostra as relaes bivariadas entre diferentes variveis, permitindo uma visualizao rpida e conveniente da distribuio dos dados e das relaes entre as variveis.'''

# sns.pairplot(df)
sns.pairplot(df, x_vars=['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number_of_Rooms', 'Avg_Area_Number_of_Bedrooms', 'Area_Population'], y_vars='Price')
```

```python
'''A funo corr()  uma funo estatstica utilizada para calcular a correlao entre duas variveis. A correlao
mede a relao entre as variveis e indica a fora e a direo dessa relao. Ela  amplamente utilizada em anlises
estatsticas para determinar o grau de associao entre as variveis e entender se existe uma relao linear entre elas.

 A funo corr() retorna um valor que varia de -1 a 1, em que:
   Um valor de 1 indica uma correlao positiva perfeita, ou seja, as variveis esto perfeitamente correlacionadas de forma positiva.
   Um valor de -1 indica uma correlao negativa perfeita, ou seja, as variveis esto perfeitamente correlacionadas de forma negativa.
   Um valor prximo a 0 indica uma correlao fraca ou inexistente entre as variveis.'''
  
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True)
```
